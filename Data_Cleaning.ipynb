Read & clean data to analyze gender & sexual orientation subgroups in relation to depression and anxiety over specific time periods during Covid.

```python
#Depression symptoms by gender
#Locate group & indicator columns to perform a groupby on df
depression_gender = mental.loc[(mental["Group"] == "By Sex")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

#Create new df for specified groupby, male vs female (gender) & depression symptoms
female_depression = depression_gender.loc[(depression_gender["Subgroup"] == "Female")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

male_depression = depression_gender.loc[(depression_gender["Subgroup"] == "Male")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]
```

```python
#Anxiety symptoms by gender
#Locate group & indicator columns to perform a groupby on df
anxiety_gender = mental.loc[(mental["Group"] == "By Sex")& \
                             (mental["Indicator"] == "Symptoms of Anxiety Disorder"),:]
#Create new df for specified groupby, male vs female (gender) & depression symptoms
female_anxiety = anxiety_gender.loc[(anxiety_gender["Subgroup"] == "Female")& \
                             (mental["Indicator"] == "Symptoms of Anxiety Disorder"),:]

male_anxiety = anxiety_gender.loc[(anxiety_gender["Subgroup"] == "Male")& \
                             (mental["Indicator"] == "Symptoms of Anxiety Disorder"),:]
```

```python
#Anxiety symptoms by sexual orientation
anxiety_sex_orientation = mental.loc[(mental["Group"] == "By Sexual orientation")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

gay_anxiety = anxiety_sex_orientation.loc[(anxiety_sex_orientation["Subgroup"] == "Gay or lesbian")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

straight_anxiety = anxiety_sex_orientation.loc[(anxiety_sex_orientation["Subgroup"] == "Straight")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

bisexual_anxiety = anxiety_sex_orientation.loc[(anxiety_sex_orientation["Subgroup"] == "Bisexual")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]
```

```python
#Depression symptoms by sexual orientation
depression_sex_orientation = mental.loc[(mental["Group"] == "By Sexual orientation")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

gay_depression = depression_sex_orientation.loc[(depression_sex_orientation["Subgroup"] == "Gay or lesbian")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

straight_depression = depression_sex_orientation.loc[(depression_sex_orientation["Subgroup"] == "Straight")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]

bisexual_depression = depression_sex_orientation.loc[(depression_sex_orientation["Subgroup"] == "Bisexual")& \
                             (mental["Indicator"] == "Symptoms of Depressive Disorder"),:]
```

```python
#Rename Columns to preapre for merging data
male_anxiety.rename(columns = {'Indicator':'Symptoms of Anxiety', 'Subgroup': 'Male Anxiety', 'Value': 'Male Anxiety Value'}, inplace = True)
female_anxiety.rename(columns = {'Indicator':'Symptoms of Anxiety','Subgroup': 'Female Anxiety', 'Value': 'Female Anxiety Value'}, inplace = True)
male_depression.rename(columns = {'Indicator':'Symptoms of Depression','Subgroup': 'Male Depression', 'Value': 'Male Depression Value'}, inplace = True)
female_depression.rename(columns = {'Indicator':'Symptoms of Depression','Subgroup': 'Female Depression', 'Value': 'Female Depression Value'}, inplace = True)
gay_anxiety.rename(columns = {'Indicator':'Symptoms of Anxiety','Subgroup': 'Gay Anxiety', 'Value': 'Gay Anxiety Value'}, inplace = True)
straight_anxiety.rename(columns = {'Indicator':'Symptoms of Anxiety','Subgroup': 'Straight Anxiety', 'Value': 'Straight Anxiety Value'}, inplace = True)
bisexual_anxiety.rename(columns = {'Indicator':'Symptoms of Anxiety','Subgroup': 'Bisexual Anxiety', 'Value': 'Bisexual Anxiety Value'}, inplace = True)

gay_depression.rename(columns = {'Indicator':'Symptoms of Depression','Subgroup': 'Gay Depression', 'Value': 'Gay Depression Value'}, inplace = True)
straight_depression.rename(columns = {'Indicator':'Symptoms of Depression','Subgroup': 'Straight Depression', 'Value': 'Straight Depression Value'}, inplace = True)
bisexual_depression.rename(columns = {'Indicator':'Symptoms of Depression','Subgroup': 'Bisexual Depression', 'Value': 'Bisexual Depression Value'}, inplace = True)
```

```python
#Filter specified columns from df 
#male & female anxiety (need date, subgroup, value,indicator)
filtered_male_anxiety = male_anxiety.iloc[:, [0,3,6,9]] 
filtered_female_anxiety = female_anxiety.iloc[:,[0,3,9]]

#male & female depression (need date, subgroup, value,indicator)
filtered_male_depression = male_depression.iloc[:,[0,3,6,9]]
filtered_female_depression = female_depression.iloc[:,[0,3,9]]

#sexual orientation anxiety (need date, subgroup, value,indicator)
filtered_gay_anxiety = gay_anxiety.iloc[:, [0,3,6,9]]
filtered_straight_anxiety = straight_anxiety.iloc[:, [0,3,9]]
filtered_bisexual_anxiety = bisexual_anxiety.iloc[:, [0,3,9]]

#sexual orientation depression (need date, subgroup, value,indicator)
filtered_gay_depression = gay_depression.iloc[:, [0,3,6,9]]
filtered_straight_depression = straight_depression.iloc[:, [0,3,9]]
filtered_bisexual_depression = bisexual_depression.iloc[:, [0,3,9]]

#Merge dataframes for final output to csv - gender
data_anxiety_gender_merge = pd.merge(filtered_male_anxiety, filtered_female_anxiety, how="left", on=["Symptoms of Anxiety"])
data_depression_gender_merge = pd.merge(filtered_male_depression, filtered_female_depression, how="left", on=["Symptoms of Depression"])
data_gender_merge = pd.merge(data_anxiety_gender_merge, data_depression_gender_merge, how="left", on=["Time Period Label"])

#Merge dataframes for final output to csv - sexual orientation
data_anxiety_pt1_merge = pd.merge(filtered_gay_anxiety, filtered_straight_anxiety, how="left", on=["Symptoms of Anxiety"])
data_anxiety_sexual_orien_merge = pd.merge(filtered_bisexual_anxiety,data_anxiety_pt1_merge, how="left", on=["Symptoms of Anxiety"])
data_depression_pt1_merge = pd.merge(filtered_gay_depression, filtered_straight_depression, how="left", on=["Symptoms of Depression"])
data_depression_sexual_orien_merge = pd.merge(filtered_bisexual_depression,data_depression_pt1_merge, how="left", on=["Symptoms of Depression"])
data_sexual_orien_merge = pd.merge(data_depression_sexual_orien_merge,data_anxiety_sexual_orien_merge, how="left", on=["Time Period Label"])
```

Vaccination Rates in States vs Mental Wellbeing Cleaning

```python
#read in csv data from datasets taken from:
#1.) Mental is based off of the CDC mental health survey found here https://data.cdc.gov/NCHS/Indicators-of-Anxiety-or-Depression-Based-on-Repor/8pt5-q6wp
#2.) Vaccination Data is taken from the CDC found here  https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-County/8xkx-amqh
mental = pd.read_csv('Resources/mental.csv')
vaccine = pd.read_csv('Resources/vaccine_trimmed.csv')
vaccine = vaccine.drop(columns =['Unnamed: 0'])
```

```python
#Filtering the data based off of the indicator which determines which symptoms are counted as a positive result from the questionaire. 
#I have chosen both anxiety and depression as my subject of interest as I more focused on overall mental unrest
mental_trimmed = mental.loc[(mental['Group'] == 'By Age') | (mental['Group'] == 'By State')].reset_index(drop=True)
mental_trimmed = mental_trimmed.loc[mental_trimmed['Indicator'] == "Symptoms of Anxiety Disorder or Depressive Disorder"]
mental_trimmed
```

```python
#After careful examination of the datasets involved in this analysis. The vaccine dataframe conveys data from all days from
#12/13/2020 to 6/13/2022. However the mental health data is taken from specific time frames and then reported as a collective.
#Due to these differences and the inability to know exactly when the mental health data was taken the vaccination data was used
#for solely the last day of a given time period in the mental health survey. This code is finding those dates and trimming them
#down until there is both mental health and vaccination status data from both dataframes.
end_dates = mental['Time Period End Date'].unique().tolist()
end_trim = end_dates[0:-1]
end_values = end_trim[21:]
end_values.remove('01/05/2021')
end_values.remove('04/13/2021')
end_values.remove('07/20/2021')
end_values.remove('11/30/2021')
end_values.remove('03/01/2022')
end_values.remove('05/31/2022')
end_values
```

```python
#This code is used to match state names to state codes. Since each dataframe is going to compared along a date from a given 
#state it is important that we have each dataframe arranged the same.
state_2 = {
    "Alabama": "AL",
    "Alaska": "AK",
    "Arizona": "AZ",
    "Arkansas": "AR",
    "California": "CA",
    "Colorado": "CO",
    "Connecticut": "CT",
    "Delaware": "DE",
    "Florida": "FL",
    "Georgia": "GA",
    "Hawaii": "HI",
    "Idaho": "ID",
    "Illinois": "IL",
    "Indiana": "IN",
    "Iowa": "IA",
    "Kansas": "KS",
    "Kentucky": "KY",
    "Louisiana": "LA",
    "Maine": "ME",
    "Maryland": "MD",
    "Massachusetts": "MA",
    "Michigan": "MI",
    "Minnesota": "MN",
    "Mississippi": "MS",
    "Missouri": "MO",
    "Montana": "MT",
    "Nebraska": "NE",
    "Nevada": "NV",
    "New Hampshire": "NH",
    "New Jersey": "NJ",
    "New Mexico": "NM",
    "New York": "NY",
    "North Carolina": "NC",
    "North Dakota": "ND",
    "Ohio": "OH",
    "Oklahoma": "OK",
    "Oregon": "OR",
    "Pennsylvania": "PA",
    "Rhode Island": "RI",
    "South Carolina": "SC",
    "South Dakota": "SD",
    "Tennessee": "TN",
    "Texas": "TX",
    "Utah": "UT",
    "Vermont": "VT",
    "Virginia": "VA",
    "Washington": "WA",
    "West Virginia": "WV",
    "Wisconsin": "WI",
    "Wyoming": "WY",
    "District of Columbia": "DC",
    "American Samoa": "AS",
    "Guam": "GU",
    "Northern Mariana Islands": "MP",
    "Puerto Rico": "PR",
    "United States Minor Outlying Islands": "UM",
    "U.S. Virgin Islands": "VI",
}

state_2 = {state: abbrev for state, abbrev in state_2.items()}
new_mental = mental_trimmed.loc[mental_trimmed['Group'] == 'By State']
new_mental = new_mental.rename(columns={'Time Period End Date': 'Date'})
new_mental = new_mental[['State', 'Value', 'Date']]
new_mental['abbrev'] = new_mental['State'].map(state_2)
new_mental = new_mental.drop(columns=['State'])
new_mental = new_mental.rename(columns={'abbrev':"State"})
new_mental
```

```python
#In this cell we are working to have the dataframe come more in line with each other. This trims the columns further until
#only the interesting parts will remain. In addition to this, having a common framework will enable easier integration.
vax_clean = vaccine
vax_clean = vax_clean.rename(columns={'Recip_State': 'State'})
new_vax = pd.DataFrame()
mental_dates = pd.DataFrame()
for column in vax_clean.columns:
    new_vax[column] = ''
for column in new_mental.columns:
    mental_dates[column] = ''
for date in end_values:
    df = new_mental.loc[new_mental['Date'] == date]
    df_2 = vax_clean.loc[vax_clean['Date'] == date]
    mental_dates = pd.concat([mental_dates, df])
    new_vax = pd.concat([new_vax, df_2])
new_vax
```

```python
#While the columns are broadly speaking in alignment there is still far to much data in the vaccine information to be parsed
#easily. As such this cell removes the bulk of that unnecessary information.
trim_vax = pd.DataFrame()
state_list = vax_clean['State'].unique().tolist()
new_state_list = state_list[0:-2]
for column in new_vax.columns:
    trim_vax[column] = ''
for date in end_values:
    date_trim = new_vax.loc[new_vax['Date'] == date]
    for state in new_state_list:
        state_trim = date_trim.loc[date_trim['State'] == state]
        complete_trim = state_trim.iloc[2:3,:]
        #print(complete_trim)
        trim_vax = pd.concat([trim_vax, complete_trim])
trim_vax = trim_vax.dropna()
trim_vax = trim_vax.drop(columns=['Recip_County']).reset_index(drop=True)
trim_vax['Date'] = pd.to_datetime(trim_vax['Date'])
trim_vax['Date'] = trim_vax['Date'].dt.strftime("%Y/%m/%d")
trim_vax = trim_vax.sort_values(by=['Date','State'])
trim_vax = trim_vax.astype({'Date': str})
trim_vax = trim_vax.reset_index(drop=True)
trim_vax = trim_vax.loc[(trim_vax['State'] != 'PR') & (trim_vax['State'] != 'VI')]
trim_vax = trim_vax.reset_index(drop=True)
trim_vax
```

```python
#Now that the vaccine information is trimmed to desired level(one date per state) the mental health dataframe is examined
mental_dates = mental_dates.reset_index(drop=True)
mental_dates
```

```python
#Further information is determiend to not be overlapped primarily that DC is not accounted for the vaccination data.
#In addition there are three dates where HI does not report their vaccination data so these are removed for the mental health
#dataframe as well.
date_trim = mental_dates.loc[mental_dates['State'] != 'DC']
date_trim['Date'] = pd.to_datetime(mental_dates['Date'])
date_trim['Date'] = date_trim['Date'].dt.strftime("%Y/%m/%d")
date_trim = date_trim.sort_values(by=['Date','State'])
date_trim = date_trim.astype({'Date': str})
date_trim = date_trim.drop([1160,1210,1260])
date_trim = date_trim.reset_index(drop=True)
date_trim
```

```python
#Appending the matching data into the vaccination dataframe
value = date_trim['Value']
trim_vax['Value'] =value
trim_vax
```
